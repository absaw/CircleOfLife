```python
import numpy as np
import pandas as pd
a={
    1:[3,6],
    2:[4,5],
    3:[4,3],
    4:[5,1],
    5:[1,0]
}

print("Initial dict",a)


```

```python
# a_df=pd.DataFrame(list(a.items()))
a_df=pd.DataFrame.from_dict(a,orient='index')
print(a_df)
```

```python
a_df.sort_values(by=[0,1],ascending=[True,False],inplace=True)
print(a_df)
```
